```python
# Taking only necessary columns
df = df[['movie_id','overview','title','genres','keywords','cast','crew','backdrop_path']]
```

```python
# Printing only 1'st row of 'genres' column
df.iloc[0,3]
```

```python
# Creating function for selecting & printing genre names from 'genre' column
import ast
def name(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
# Applying created function on 'genres' column
df['genres']=df['genres'].apply(name)
```

```python
# Printing dataset
df.head()
```

```python
# Printing only 1'st row of 'cast' column
df.iloc[0,5]
```

```python
# Creating function for selecting & printing first 3 actor names from 'cast' column
import ast
def name2(obj):
  L=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
# Applying created function on 'cast' column
df['cast']=df['cast'].apply(name2)
```

```python
df.head()
```

```python
# Printing only 1'st row of 'crew' column
df.iloc[0,6]
```

```python
# Creating function for selecting & printing director name from 'crew' column
def director_name(obj):
  l= []
  counter=0
  for i in ast.literal_eval(obj):
    if counter !=1:
      l.append(i['name'])
      counter+=1
  return l
```

```python
# Applying created function on 'crew' column
df['crew'] = df['crew'].apply(director_name)
```

```python
# Applying split fuction on overview column (Converting overview column in list)
df['overview'] = df['overview'].apply(lambda x :x.split())
```

```python
# Applying literal_eval( 'remake' ==> remake)
#evaluates a string as a literal expression, and returns the result
df['keywords'] = df['keywords'].apply(ast.literal_eval)
```

```python
# Printing dataframe
df.head()
```

```python
# To avoid creating different variations of one name (eg. Aditya Patil ==> Aditya & Patil) this step will remove space between two words
df['cast'] = df['cast'].apply(lambda x: [i.replace(' ','')for i in x])
df['genres'] = df['genres'].apply(lambda x: [i.replace(' ','')for i in x])
df['overview'] = df['overview'].apply(lambda x: [i.replace(' ','')for i in x])
df['keywords'] = df['keywords'].apply(lambda x: [i.replace(' ','')for i in x])
df['crew'] = df['crew'].apply(lambda x: [i.replace(' ','')for i in x])
```

```python
# Merging all columns into one column 'tags'
df['tags'] = df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
# Printing dataframe
df.head()
```

```python
# Removing other columns
df=df[['movie_id','title','tags']]
```

```python
#convert tags column in string
df['tags'] = df['tags'].apply(lambda x: ' '.join(x))
```

```python
# Converting 'tags' column in lower case
df['tags'] = df['tags'].apply(lambda x: x.lower() )
```

```python
df.head()
```

```python
# Importing NLTK
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Applying stemming to deal with words like -['ability','abilities']
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
# Removing stopwords
def remove_stopwords(text):
  y=[]
  stop_words = set(stopwords.words('english'))
  words = word_tokenize(text)
  filtered_words = [y.append(words) for words in  words if words.lower() not in stop_words]
  return ' '.join(y)
```

```python
# Applying stopwords on 'tags' column
df['tags'] = df['tags'].apply(remove_stopwords)
```

#model building

```python
# Countvectorizer converts datapoints into vectors and count their no. of occurence
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
#Output is given in sparse form so, this step
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors[0]
```

```python
# Most frequent words
feat_dict=cv.vocabulary_.keys()
sorted(feat_dict)
```

```python
# Cosine similarity is used to find diastanve betewwn vectors

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
df[df['title']=='Pathaan'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
# Final functiom to get movovie remmendation
def recommend_movie(movies):
  movie_index = df[df['title']==movies].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(df.iloc[i[0]].title)
```

# Get Your Movie Recommended

```python
recommend_movie('Pathaan')
```
